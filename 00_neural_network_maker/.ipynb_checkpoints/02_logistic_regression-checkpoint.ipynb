# ロジスティック回帰による2クラス分類

単純パーセプトロンではステップ関数を用いてある入力ベクトルがどちらのクラスに属するのかを明示的に出力しました。<br>
しかしながら実際の問題では入力に対してどちらのクラスに属するかの確率を出力した方が有用な状況があります。<br>
（癌の検知においては51%の確率で癌でないケースと99%の確率で癌でないケースを分けれた方が良いでしょう)<br>
そこで確率を出力するためにステップ関数 $f(x)$ の代わりに次のような関数$\sigma(x)$を用いる場合があります
$$
\sigma(x)=\frac{1}{1+e^{-x}}
$$
この形で表される関数$\sigma$のことを（ロジスティック）シグモイド関数と呼びます

````python
import numpy as np
from scipy.special import erf
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
#シグモイド関数
def sigmoid(x):
    return (1/(1+np.exp(-t)))
t = np.linspace(-10, 10, 1000)
y = sigmoid(t)
plt.plot(t, y)
plt.show()
````

ここで、横軸に取られたtは入力$\bf x$に対して$\   t={\bf w}^T{\bf x}+b\   $で求まり、決定面からどれほど、どちら側に離れているのかを表す値となっています。例えばある入力ベクトルに対して$t=-0.33$と出たとしましょう。ここまでは単純パーセプトロンで行った処理と同じです。
しかし活性化関数の違いによりそれぞれの出力は次のようになります<br><br>
活性化関数にステップ関数を用いた場合<br>
$$
f(-0.33)=(-0.33\geq0)\ ?\ 1:0=0
$$
活性化関数に（ロジスティック）シグモイド関数を用いた場合
$$
\sigma(-0.33)=\frac{1}{1+e^{-(-0.33)}}\risingdotseq0.42
$$

ここで、0.42の表す意味について考えると0.42の確率でクラス1に属している（あるいは0.58の確率でクラス2に属している)ということになります。<br><br>
***
順序が前後しますがこれからは確率を出力する活性化関数をどのように定めるべきなのかについて考えていきます。
まず、入力されるtに対してですが$\   t={\bf w}^T{\bf x}+b\   $のbによってt=0のとき、（予測される）決定面上に入力ベクトルが位置していることがわかりますので
t=0の時、活性化関数は0.5を出力する必要があります。加えて、何かしらの確率分布に対する累積分布関数になっている必要があります。ここでいう何かしらの確率分布とは**決定面がどこであるかに対する予測**になります。決定面に対する予測を正規分布と置きパラメータを変化させてみましょう

````python
#ガウス分布
def gauss(x, sigma):
    return (1/((2*np.pi*sigma)**0.5))*np.exp(-((x**2)/(2*sigma)))

#ガウス分布に対する累積分布関数
def Gauss(x, sigma):
    return (1.0 + erf(x / np.sqrt(2.0*sigma))) / 2.0

t = np.linspace(-10, 10, 1000)
y = [gauss(t, 1), gauss(t, 4.0), gauss(t, 16.0)]
Y = [Gauss(t, 1), Gauss(t, 4.0), Gauss(t, 16.0)]
fig, (L, R) = plt.subplots(ncols=2, figsize=(10,4))
L.plot(t, y[0])
L.plot(t, y[1])
L.plot(t, y[2])
R.plot(t, Y[0])
R.plot(t, Y[1])
R.plot(t, Y[2])
fig.show()

#期待値 μ=0 分散 σ^2=[1.0, 4.0, 16.0] におけるガウス分布と累積分布関数
````

このグラフから読み取れることは決定面に対する確信度合いが強ければ強いほど、つまり分散が小さければ小さいほど、累積分布関数（活性化関数）が出力する確率も確信的になっていくということです。このように正規分布の累積分布関数を用いるモデルのことをプロビット回帰と呼びます。また決定面に対して絶対の確信がある時（$\ \lim_{\sigma^2 \to +0}\ $）の時,この活性化関数はステップ関数となります。<br>
***
これまでの議論により、決定面に対する確率分布を求めてからその累積分布関数（活性化関数）を出すことが理想的であることがわかりました。
<br>しかしながら実際には計算量の問題からあまり使われることはありません。そこで使用されるのが(ロジスティック)シグモイド関数となります。
累積分布関数としてシグモイド関数を見ると

````python
#シグモイド関数の微分
def sigmoid_dif(x):
    return sigmoid(x)*(1-sigmoid(x))

t = np.linspace(-10, 10, 1000)
y = [sigmoid_dif(t), gauss(t, 4.0)]
Y = [sigmoid(t), Gauss(t, 4.0)]
fig, (L, R) = plt.subplots(ncols=2, figsize=(10,4))
L.plot(t, y[0])
L.plot(t, y[1])
R.plot(t, Y[0])
R.plot(t, Y[1])
fig.show()
````

$\sigma^2=4.0$の正規分布と並べて見ると非常に近い活性化関数になっていることがわかります。またシグモイド関数の微分は$$\sigma'(x)=\sigma(x)(1-\sigma(x))$$となります($\ (\frac{1}{g(x)})'=\frac{g'(x)}{g^2(x)}\ $から簡単に導けます）。後にこの性質が計算量削減につながります。<br><br><br>
では実際にシグモイド関数を用いて確率を出力するモデルを作っていきます。${\bf x}$が$C=\{0, 1\}$のうち1に属する確率は$p(C＝１|{\bf x})=\sigma({\bf w}^T{\bf x}+b)$となり、$y=\sigma({\bf w}^T{\bf x}+b)$として置くと${\bf x}$がクラス$t$に属する確率を次のように表せます。<br><br>
$$
p(C=t|{\bf x})=y^t(1-y)^t
$$

この確率モデルについて尤度関数を最大化していきます
尤度関数は
$$
L(y) = \prod_{n=1}^{N} y_{n}^{t_n}(1-y_{n})^{1-t_n}
$$
この尤度関数を最大にする$\bf w$を求めることが目的なのですがこれは誤差関数$E(y)=-\log L(y)$を最小化する問題として置き換えることができます。<br>
誤差関数の対数をとって微分すると
$$
E(y) = -\sum_{n=1}^N  \{ t_n \log y_n +(1-t_n) \log (1-y_n)\}\\
\frac{E(y)}{\delta y} \frac{\delta y}{\delta {\bf w}}=-\sum_{n=1}^N(\frac{t_n}{y_n}-\frac{1-t_n}{1-y_n})\frac{\delta y}{\delta {\bf w}}\\
$$
$y=\sigma({\bf w}^T{\bf x}+b)$なので
$$
\begin{eqnarray*}
\frac{E(y)}{\delta y} \frac{\delta y}{\delta {\bf w}}&=&-\sum_{n=1}^N(\frac{t_n}{y_n}-\frac{1-t_n}{1-y_n})y_n(1-y_n){\bf x_n}\\
&=&-\sum_{n=1}^N(t_n(1-y_n)-y_n(1-t_n)){\bf x}_n\\
&=&-\sum_{n=1}^N(t_n-y_n){\bf x}_n
\end{eqnarray*}
$$
同じようにして$b$について微分すると
$$
\frac{\delta E(y)}{\delta y}\frac{\delta y}{\delta b}=-\sum_{n=1}^{N}(t_n-y_n)
$$
<br>
となります。式をみてわかるように確率分布モデルを$\ y^t(1-y)^{1-t}\ $とする場合には活性化関数を$\ \frac{1}{1-e^x}\ $とすると誤差計算が楽にできます。<br>
***
誤差関数の微分を求めることができたので$\bf w$と$b$の更新式がそれぞれ
$$
{\bf w}^{(k+1)}={\bf w}^{(k)}-\eta \sum_{n=1}^{N}(y_n-t_n){\bf x}_n\\
b^{(k+1)}=b^{(k)}-\eta \sum_{n=1}^{N}(y_n-t_n)
$$
となります。この時の$\eta$を学習の際のパラメータの変動幅を決定する係数となっているので、**決定係数**と呼びます。
また更新式を見ると一度の学習をするために毎度、全データに対する誤差を計算する必要があることがわかります。<br>
データの総数(N)とデータの次元数(D)が小さな数の場合には問題ありませんがこれらが大きくなるとコンピュータ上ではメモリ不足が生じたり計算時間が膨大になるといった問題が生じます。<br>
そこでデータを一つずつランダムに抽出してパラメータを更新する**確率的勾配降下法**や、データをランダムにM個抽出しパラメータを更新する**ミニバッチ勾配降下法**があります。<br>
更新式はそれぞれ以下のようになります。<br>
***
確率的勾配降下法
$$
{\bf w}^{(k+1)}={\bf w}^{(k)}-\eta (y_n-t_n){\bf x}_n\\
b^{(k+1)}=b^{(k)}-\eta (y_n-t_n)
$$
***
ミニバッチ勾配降下法
$$
{\bf w}^{(k+1)}={\bf w}^{(k)}-\eta \sum_{n=1}^{M}(y_n-t_n){\bf x}_n\\
b^{(k+1)}=b^{(k)}-\eta \sum_{n=1}^{M}(y_n-t_n)
$$
これらの方法を用いて学習していくのですが一般的に一度の更新でパラメータが収束することはないのでデータ全体に対してデータ選択の偏りが出ないようにシャッフルしながら繰り返し学習します。この反復回数のことを**エポック**といいます。

````python
class LogisticRegression(object):
    def __init__(self, input_dim):
        self.input_dim = input_dim
        self.w = np.random.normal(size=(input_dim, ))
        self.b = 0
        
    def __call__(self, x):
        return self.forward(x)
    
    def forward(self, x):
        return self.sigmoid(np.matmul(x, self.w)+self.b)
    
    def compute_gradients(self, x, t):
        y = self.forward(x)
        delta = y-t
        dw = np.matmul(x.T, delta)
        db = np.matmul(np.ones(x.shape[0]), delta)
        return dw, db
    
    def compute_loss(self, t, y):
        return (-t*np.log(y) - (1-t)*np.log(1-y)).sum()

    def train_step(self, x, t):
        dw, db = self.compute_gradients(x, t)
        model.w = self.w - 0.1*dw
        model.b = self.b - 0.1*db
        loss = self.compute_loss(t, self(x))
        return loss

    def sigmoid(self, x):
        return 1/(1+np.exp(-x))
````

````python
#データの準備 0Rゲートの学習
x = np.array([[0,0], [0,1], [1,0], [1,1]])
t = np.array([0, 1, 1, 1])

#xor回路は線形分離不可能なので識別できません
#試しにやってみてください
#t = np.array([0, 1, 1, 0])
````

````python
#モデルの構築
model = LogisticRegression(input_dim=2)
````

````python
#モデルの学習
epochs = 100
for epoch in range(epochs):
    train_loss = model.train_step(x, t)#バッチ学習
    if epoch%10 == 0 or epoch == epochs-1:
        print('epoch: {}, loss: {:.3f}'.format(
            epoch+1,
            train_loss
        ))
````

````python
#モデルの評価
for input in x:
    print('{} => {:.3f}'.format(input, model(input)))
````
