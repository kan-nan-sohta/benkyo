# 様々な活性化関数
## 勾配消失問題
これまで活性化関数としてシグモイド関数 $$\sigma(x)=\frac{1}{1+e^{-x}}$$
を使用してきました。この活性化関数の微分値は $$\sigma'(x)=\sigma(x)(1-\sigma(x))$$
でありこの関数の最大値は

````python
import numpy as np
import copy
from scipy.special import erf
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
def sigmoid(x):
    return (1/(1+np.exp(-t)))
t = np.linspace(-10, 10, 1000)
y = sigmoid(t)
y_ = sigmoid(t) * (1-sigmoid(t))
plt.plot(t, y)
plt.plot(t, y_)
plt.show()
````

0の時に最大となり、その最大値が0.25であることがわかります。<br>
ロジスティック回帰ではシグモイド関数を通した出力との誤差に対してその微分値をパラメータの更新の際の変化量とします。<br>
多層になるとその誤差を伝播していくことになるのでn層のネットワークの最奥の層では学習率が$0.25^n$倍されてしまいます。<br>
これでは効果的な学習を実現できません。よって微分しても値が小さくならない活性化関数を与えることで解決するアプローチがあります。<br>
これからはそのアプローチに使われるいくつかの活性化関数を紹介していきます。
***
## 双曲線正接関数(tanh)
ハイパボリックタンジェント関数とも呼ばれる関数で $$ tanh(x)= \frac{e^x - e^{-x}}{e^x + e^{-x}} $$ 
で表されます。この関数の微分値は　$$ tanh'(x)= \frac{4}{(e^x + e^{-x})^2} $$
となります。実際にこの関数の概形を表します。

````python
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

def dtanh(x):
    return (4)/((np.exp(x)+np.exp(-x))**2)
t = np.linspace(-10, 10, 1000)
y = tanh(t)
y_ = dtanh(t)
plt.plot(t, y)
plt.plot(t, y_)
plt.show()
````

微分した際の最大値が1になっているのが確認でき、勾配が消失しにくくなっていることがわかると思います。<br>
値域がシグモイド関数と違い -1<tanh(x)<1 となっていることに注意してください。
Keras, Tensorflow, Pytorch　それぞれでの実装は次のようになります。
```python
#Keras
model.add(Dense(hidden_dim, activation='tanh')
#Tensorflow
self.l1 = Dense(hidden_dim, activation='tanh')
#Pytorch
self.l1 = nn.Linear(input_dim, hidden_dim)
self.a1 = nn.tanh()
```

## Relu
***
ランプ関数とも呼ばれる関数です。双曲線正接関数では依然として入力次元や隠れ層が増えると勾配が消失してしまうという問題がありました。<br>その問題を解消するべく考案された活性化関数です。<br>
式で表すと　$$ f(x) = max(0, x) $$
微分は　$$ f'(x) = \begin{cases} 1 (x>0) \\ 0 (x\leq0)\end{cases} $$
となります。実際にはx=0で微分不可能ですがこのように定義します。

````python
def relu(x):
    l = np.zeros(len(x))
    return np.maximum(x, l)

def drelu(x):
    l = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] > 0:
            l[i] = 1
    return l
t = np.linspace(-10, 10, 1000)
y = relu(t)
y_ = drelu(t)
plt.plot(t, y)
plt.plot(t, y_)
plt.show()
````

xが大きい値の場合でも勾配が消失することがなく、計算自体も単純でステップ単位でも学習の高速化が期待できます。<br>
学習率を大きくしすぎたりして初期のステップでニューロンが不活性になるとそれ以降の勾配が0になり、それ以降学習が行われないことがあるという欠点もあります。
```python
#Keras
model.add(Dense(hidden_dim, activation='relu')
#Tensorflow
self.l1 = Dense(hidden_dim, activation='relu')
#Pytorch
self.l1 = nn.Linear(input_dim, hidden_dim)
self.a1 = nn.ReLU()
```

## Leaky Relu
***
Reluから派生して考えられた活性化関数です。一度不活性になると学習が進まない問題を解決するべく考えられた関数ですが実際には効果があるときとない時がまちまちだそうです。
式で表すと　$$ f(x) = max(ax, x) (0<a<1)$$
微分は　$$ f'(x) = \begin{cases} 1 (x>0) \\ a (x\leq0)\end{cases} $$

````python
def lrelu(x):
    l = copy.deepcopy(x)*0.1
    return np.maximum(x, l)

def dlrelu(x):
    l = np.ones(len(x))*0.1
    for i in range(len(x)):
        if x[i] > 0:
            l[i] = 1
    return l
t = np.linspace(-10, 10, 1000)
y = lrelu(t)
y_ = dlrelu(t)
plt.plot(t, y)
plt.plot(t, y_)
plt.show()
````

xがマイナスになったときでも勾配が失われなず学習が可能であることがわかると思います。<br>
実装に関してはKeras Tensorflowで少し変化がありDenseにLeakyReluがないため別途、活性化層をインポートする必要があることに注意しましょう
```python
#Keras
model.add(Dense(hidden_dim))
model.add(LeakyReLU(0.1))
#Tensorflow
self.l1 = Dense(hidden_dim)
self.a1 = LeakyReLU(0.1)
#Pytorch
self.l1 = nn.Linear(input_dim, hidden_dim)
self.a1 = nn.LeakyReLU()
```

## Swish
***
2017年に提案されたReLUベースでない活性化関数としてSwishがあります。次の式で定義されます。
$$ f(x) = x*\sigma(\beta x), \space\beta > 0$$
微分は
$$ f(x) = \beta f(x) + \sigma(\beta x)(1-\beta f(x)) $$
また$\beta$を無限大に飛ばすとReLUに近づくことが判明しています。

````python
def swish(beta, x):
    l = copy.deepcopy(x)*beta
    for i in range(len(x)):
        l[i] = x[i] * (1/(1+np.exp(-l[i])))
    return l

def dswish(beta, x):
    l = copy.deepcopy(x)*beta
    for i in range(len(x)):
        l[i] = (1/(1+np.exp(-l[i]))) + l[i]*(1/(1+np.exp(-l[i])))*(1-(1/(1+np.exp(-l[i]))))
    return l


fig, axes = plt.subplots(ncols=2, figsize=(12, 4))
t = np.linspace(-10, 10, 1000)

y1 = swish(0.1, t)
y1_ = dswish(0.1, t)
axes[0].plot(t, y1)
axes[1].plot(t, y1_)

y2 = swish(1.0, t)
y2_ = dswish(1.0, t)
axes[0].plot(t, y2)
axes[1].plot(t, y2_)

y3 = swish(10.0, t)
y3_ = dswish(10.0, t)
axes[0].plot(t, y3)
axes[1].plot(t, y3_)
fig.show()
````

連続性を保ったままLeRUに似た関数を実装することができました。いくつかの実験においてReLUを超える成果を残してますが、基本的にはReLUが用いられるようです。<br>
実装では関数から定義する必要があります。

活性化関数にできるだけ勾配が消失しないような活性化関数を使用することで学習を効率的に行いたいというのなら<br>
最も単純で勾配の消失しない線形関数を使用すればいいじゃん！<br>
## A.多層にする意味がなくなります
活性化関数に線形関数を用いると隠れ層の積み重ねが全て一つの層で実現できてしまいます。以下の多層モデルについて考えると
<img src="pics/linear.png" width="500">
$$\begin{eqnarray}
sigmoid(x''+y'') &=& sigmoid(a'x'+c'y'+b'x'+d'y')\\
&=& sigmoid(a'(ax+cy)+c'(bx+dy)+b'(ax+cy)+d'(bx+dy))\\
&=& sigmoid(a'ax+a'cy+c'bx+c'dy+b'ax+b'cy+d'bx+d'dy)\\
&=& sigmoid((a'a+c'b+b'a+d'b)x + (a'c+c'd+b'c+d'd)y)
\end{eqnarray}
$$
<img src="pics/linear2.png" width="500">
となるので多層にする必要がありません。人間の脳内のニューロンが受け取った電位がある値を超えたときに発火し次のニューロンに伝達するように<br>
大きすぎたり、小さすぎる入力値に対してはある程度出力値を丸めたり、ある範囲の入力値を無視したりする必要があります。
